```python
# Lista de invitados del primer ejemplo
invitados = ["Ana", "Beto", "Carla"]

# Usando range() y len() para obtener la posición (índice) len nos indica el largo de cada nombre y ordena las posiciones según la longitud.
for i in range(len(invitados)):
    invitado = invitados[i]
    # Sumamos 1 a 'i' para que la posición sea más natural para humanos (1, 2, 3...)
    print(f"Invitado número {i + 1}: {invitado}")
```

```python
# Definimos la lista de libros con los libros de Marzo 2026
libros_top = [
    {"titulo": "La asistenta", "autor": "Freida McFadden", "busquedas": "Muy Alta"},
    {"titulo": "Comerás flores", "autor": "Lucía Solla Sobral", "busquedas": "En aumento"},
    {"titulo": "La ciudad de las luces muertas", "autor": "David Uclés", "busquedas": "Alta"},
    {"titulo": "Innamorata", "autor": "Ava Reid", "busquedas": "Tendencia"}
]

# Definimos la función
def mostrar_tendencias(lista_libros):
    print("📚 RANKING DE BÚSQUEDAS: MARZO 2026")

    # Usamos el bucle for con enumerate para la posición
    for i, libro in enumerate(lista_libros, start=1):
        # Extraemos los datos de cada libro
        titulo = libro["titulo"]
        autor = libro["autor"]
        interes = libro["busquedas"]

        # Imprimimos de forma organizada
        print(f"{i}.{titulo.upper()}")
        print(f"Autor: {autor}")
        print(f"Interés: {interes}")
        print(" " *1)

# Llamamos a la función para que se ejecute
mostrar_tendencias(libros_top)
```

```python
def mostrar_libros_vendidos(lista_libros):
    print(f"📚 LOS MÁS VENDIDOS DEL 3 DE MARZO 2026")

# El bucle for recorre la lista y enumerate nos da la posición
    for posicion, libro in enumerate(lista_libros, start=1):
        print(f"{posicion}. {libro['titulo']} - {libro['autor']}")

# Cada día los datos dentro de esta lista cambiarán automáticamente
libros_del_dia = [
    {"titulo": "La asistenta", "autor": "Freida McFadden"},
    {"titulo": "Comerás flores", "autor": "Lucía Solla Sobral"},
    {"titulo": "La ciudad de las luces muertas", "autor": "David Uclés"}
]

# Llamada a la función
mostrar_libros_vendidos(libros_del_dia)
```
